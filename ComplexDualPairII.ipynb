```python
from typing import List, Tuple, Optional
def is_valid_mu(mu: List[int]) -> Optional[Tuple[List[int], List[int]]]:
    """
    Check if μ satisfies the required conditions and return a and b if valid.

    Args:
    mu (List[int]): A list of integers representing μ

    Returns:
    Optional[Tuple[List[int], List[int]]]: A tuple (a, b) if μ is valid, None otherwise
    where a is the positive part and b is the non-positive part of μ
    """
    # Check if μ is decreasing (non-increasing)
    if not all(mu[i] >= mu[i+1] for i in range(len(mu)-1)):
        return None
    # Use filter to separate positive and non-positive parts
    a = list(filter(lambda x: x > 0, mu))
    b = list(filter(lambda x: x < 0, mu))
    return a, b

def rho(a : int):
    return list(range(a-1, -a, -2))
print(rho(3))

def lift_parameter(mu: List[int], nu: List[int], n: int) ->Optional[Tuple[List[int], [List[int]]]]:
    """
    Lift the parameters μ and ν as described in the theorem.

    Args:
    mu: List of integers representing μ
    nu: List of integers representing ν
    n: Integer representing the dimension of the target GL(n)

    Returns:
    A tuple (μ2, ν2) where μ2 and ν2 are the lifted parameters,
    or (None, None) if lifting is not possible.
    """
    m = len(mu) # Length of mu
    if len(nu) != m:
        raise ValueError("Length of ν must be equal to length of mu")

    result = is_valid_mu(mu)
    if result is None:
        raise ValueError("Invalid μ") 
    
    a, b = result
    k = len(a)
    l = len(b)

    if m <= n:
        # Case (a): m ≤ n
        mu2 = a + [0] * (n- k-l) + b
        nu2 = nu[:m-l] + rho(n-m) + nu[m-l:]
        assert(len(mu2) == n and len(nu2) == n)
        return mu2, nu2
    else:
        # Case (b): m > n
        print("rho(m-n)",rho(m-n))
        c = nu[k:-l]
        for x in rho(m-n):
            try:
                c.remove(x)
            except ValueError:
                pass
        print("c",c)
        if len(c) != n-k-l:
            return None # Lifting does not exist
        mu2 = a + [0]*(n-k-l) + b
        nu2 = nu[:k] + c + nu[m-l:]
        return mu2, nu2
def trivial_repn(n : int):
    return lift_parameter([],[],n)

def test_lift_parameter(mu: List[int], nu: List[int], n: int) -> None:
    """
    Test the lift_parameters function with given inputs and print the results.

    Args:
    mu (List[int]): List of integers representing μ
    nu (List[int]): List of integers representing ν
    n (int): Target dimension
    """
    print(f"Original (μ, ν): ({mu}, {nu})")
    print(f"Target dimension n: {n}")

    res = lift_parameter(mu, nu, n)

    if res:
        print(f"Lifted (μ, ν): ({res[0]}, {res[1]})")
    else:
        print("Lifting does not exist.")
    return res
    
```

```python
mu1 = [5, 3, 2, 0, -1, -3]
nu1 = [1, 2, 3, 4, 5, 6]

res = test_lift_parameter(mu1, nu1,8)
test_lift_parameter(res[0], res[1],len(mu1))
res=test_lift_parameter(mu1, nu1,10)
test_lift_parameter(res[0], res[1],len(mu1))
```
